```python
path = r'E:\Non Medal Clips'
prefix = 'Valorant'
prefix = prefix+'_'
```

```python
format_data = '%b-%d-%Y %I.%M.%S%p'
for file in os.listdir(path):
    if not os.path.isfile(os.path.join(path, file)):
        continue
    #correct_format = str(datetime.datetime.fromtimestamp(os.path.getmtime(os.path.join(path,file))).strftime(format_data))
    if prefix != file[:len(prefix)]:
        correct_format = prefix+file
        os.rename(os.path.join(path,file), os.path.join(path,correct_format))
```

```python
videos = []
for root, _, files in os.walk(path):
    for file in files:
        if file.endswith('mp4'):
            videos.append(os.path.join(root, file))
        #prefix = os.path.join(root, file).split('_')[0].split('\\')[-1]
        #images_loc = os.path.join(root, 'images')
        #if not os.path.exists(images_loc):
            #os.mkdir(images_loc)
        #os.system(f'ffmpeg -i "{os.path.join(root, file)}" -start_number {counter} -qscale:v 2 "{os.path.join(images_loc, prefix+"_%07d.jpg")}"')
        #counter_list = [x for x in os.listdir(images_loc) if os.path.isfile(os.path.join(images_loc, x))]
        #counter = len(counter_list)
```

```python
tracker = 0
subcounter = 0
current = ''
counter = 0
time1 = time.time()
for video in videos:
    tracker += 1
    prefix = video.split('_')[0].split('\\')[-1]
    roott = ''.join(video.rsplit('\\', 1)[:-1])
    if roott == current:
        subcounter += 1
    else:
        subcounter = 0
        current = roott
    print(f'Working on video {tracker} out of {len(videos)} ({subcounter+1} out of {len(os.listdir(roott))} in {prefix})')
    #print(f'root = {roott}')
    images_loc = r'E:\Non Medal Clips\images'
    #print(f'images loc = {os.path.join(images_loc, prefix)}')
    counter_list = [x for x in os.listdir(os.path.join(images_loc, prefix)) if os.path.isfile(os.path.join(images_loc, prefix, x))]
    counter = len(counter_list)
    #print(f'counter = {counter}')
    os.system(f'ffmpeg -i "{video}" -vf fps=10 -start_number {counter+1} -qscale:v 2 "{os.path.join(images_loc, prefix, prefix+"_%07d.jpg")}"')
time2 = time.time()
print(f'Completed in {time2-time1} seconds.')
```

```python
mp4_count = 0
for root, _, files in os.walk(path):
    for file in files:
        suffix = os.path.join(root, file).split('_')[-1].split('.')
        images_loc = os.path.join(root, 'images')
        if not os.path.exists(images_loc):
            os.mkdir(images_loc)
        for val in suffix:
            if val == 'mp4':
                mp4_count += 1
        if mp4_count == 2:
            os.rename(os.path.join(root, file), os.path.join(root, file).rsplit('.', 1)[0])
        mp4_count = 0
```
